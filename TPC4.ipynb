```python
lista1 = []

def criarlista():
    from random import randrange
    lista1.clear()
    s = 1
    N = int(input("Escolhe o comprimento da lista"))
    while s <= N:
        e = randrange(1, 101)
        lista1.append(e)
        s = s + 1
    print("Lista criada:", lista1)

def lelista():
    lista1.clear()
    x = 1
    H = int(input("Escolhe o comprimento da lista"))
    while x <= H:
      n = int(input("Introduza um número"))
      x = x +1
      lista1.append(n)
    print("Lista criada:", lista1)

def soma():
    soma = 0
    for i in lista1:
       soma = soma + i
    print ("A soma dos elementos da lista é igual a", soma,".")

def media():
    soma = 0
    for i in lista1:
        soma = soma + i
    media = soma / len(lista1)
    print("A média dos elementos da lista é igual a", media,".")

def maior():
    maior = 0
    for i in lista1:
        if i > maior:
            maior = i
    print("O maior elemento da lista é o número", maior,".")

def menor():
    menor = lista1[0]
    for i in lista1:
        if i < menor:
            menor = i
    print ("O menor elemento da lista é o número", menor,".")

def estaOrdenadaCrescente():
    if all(lista1[i] <= lista1[i + 1] for i in range(len(lista1) - 1)):
        print("Sim")
    else:
        print("Não")

def estaOrdenadaDecrescente():
    if all(lista1[i] >= lista1[i + 1] for i in range(len(lista1) - 1)):
        print("Sim")
    else:
        print("Não")

def procurarelemento():
    f = int(input("Introduz o número que pretendes procurar"))
    x = 0
    for i in lista1:
        if f == i:
            while x < len(lista1):
               if f == lista1[x]:
                print("O número", f, "está na posição", x,".")
               x = x + 1
    if f != i:
       print("-1")
    
while True:
     print("""Menu:
      (1) Criar lista
      (2) Ler lista
      (3) Soma
      (4) Média
      (5) Maior
      (6) Menor
      (7) estaOrdenada por ordem crescente
      (8) estaOrdenada por ordem decrescente
      (9) Procura um elemento
      (0) Sair""")

     opção = int(input("Escolhe uma opção, introduzindo o número correspondente"))

     if opção == 1:
        criarlista()
     
     elif opção == 2:
        lelista()
      
     elif opção == 3:
        soma()

     elif opção == 4:
        media()

     elif opção == 5:
         maior()

     elif opção == 6:
         menor()
      
     elif opção == 7:
         estaOrdenadaCrescente()
     
     elif opção == 8:
         estaOrdenadaDecrescente()
    
     elif opção == 9:
         procurarelemento()
     
     elif opção == 0:
         print("Lista criada:", lista1)
         print ("Programa Finalizado")
         break
     else:
         print("Opção inválida. Tente novamente.")        
```
