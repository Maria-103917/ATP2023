```python
cinema = []
sala1 = [150, [], "Culpa das estrelas"]
sala2 = [200, [], "Jumangi"]
sala3 = [100, [], "Talk to me"]
cinema = [sala1, sala2, sala3]

def listar(cinema):
    print("--------Cinema--------")
    for sala in cinema:
        print("Filme em exibição:", sala[2])    
    print("----------------------")

def disponível(cinema,filme,lugar):
    cond = False
    for sala in cinema:
        if sala[2]==filme:
            if lugar not in sala [1]:
                cond = True
    return cond

def vendebilhete(cinema, filme, lugar):
    for sala in cinema:
        if sala[2]==filme:
            if lugar not in sala [1]:
                sala[1].append(lugar)
            else:
                print("--------------------")
                print("O lugar está ocupado")
                print("--------------------")
    return cinema

def listardisponibilidades(cinema):
    for sala in cinema:
        print("----------------------------------------------------------------------")
        print("Filme em exibição:", sala[2], "| Lugares disponíveis:", sala[0]-len(sala[1]))
        print("----------------------------------------------------------------------")

def inserirSala(cinema, sala):
    if sala not in cinema:
        cinema.append(sala)

condição = True
while condição:
    
    print("Menu de Opções:")
    print("1. Listar filmes em exibição")
    print("2. Verificar disponibilidade de um lugar")
    print("3. Venda de bilhete")
    print("4. Listar disponibilidades das salas")
    print("5. Inserir uma nova sala")
    print("6. Sair")

    escolha = int(input("Escolha uma opção: "))

    if escolha == 1:
        listar(cinema)

    elif escolha == 2:
        filme = input("Nome do filme: ")
        lugar = int(input("Número do lugar: "))
        print(disponível(cinema,filme,lugar))

    elif escolha == 3:
        filme = input("Nome do filme: ")
        lugar = int(input("Número do lugar: "))
        print(vendebilhete(cinema, filme, lugar))

    elif escolha == 4:
        listardisponibilidades(cinema)

    elif escolha == 5:
        salanome = input("Nome da nova sala:")
        nlugares = int(input("Número de lugares disponíveis:"))
        filme = input("Nome do filme em exibição:")
        sala = [nlugares, [], filme]
        inserirSala(cinema, sala)

    elif escolha == 6:
        print("-------------------")
        print("A sair da aplicação")
        print("-------------------")
        condição = False
    
    else:
        print("----------------------------------------------------")
        print("Opção inválida. Por favor, escolha uma opção válida.")
        print("----------------------------------------------------")
```
