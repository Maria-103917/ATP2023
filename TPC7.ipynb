```python
#Aquecimento

def aquecimento():
    file = open("polinómios.txt", "w")
    p1 = [(1,7), (-3.7,4), (5,3), (-88,0)]
    p2 = [(2,2), (17,0)]
    p3 = [(7,4), (6,3), (-1,2), (43,0)]
    p4 = [(1,4)]
    p5 = [(1,2)]
    p6 = [(1,3)]
    file.write(str(p1) + "\n" + str(p2) + "\n" + str(p3) + "\n" + str(p4) + "\n" + str(p5) +"\n" + str(p6))
    file.close()
 
#Derivada de um polinómio

def derivadapolinomio(p):
    derivada = []
    for coeficiente, grau in p:
        if int(grau) == 0:
            coefderivado = 0
            grauderivado = 0
            derivada.append((coefderivado, grauderivado))
        else:
            coefderivado = int(coeficiente) * int(grau)
            grauderivado = int(grau) - 1
            derivada.append((coefderivado, grauderivado))
    return derivada

#Soma de polinómios

def somapolinomios(p1, p2):
    soma = []
    x = 0
    y = 0
    while x < len(p1) and y < len(p2):
        coef1, grau1 = p1[x]
        coef2, grau2 = p2[y]
        if grau1 == grau2:
            coef = coef1 + coef2
            grau = grau1
            soma.append((coef, grau))
            x = x + 1
            y = y + 1
        elif grau1 > grau2:
            soma.append((coef1, grau1))
            x = x +1
        else:
            soma.append((coef2, grau2))
            y = y + 1

    while x < len(p1):
        coef1, grau1 = p1[x]
        soma.append((coef1, grau1))
        x = x + 1 

    while y < len(p2):
        coef2, grau2 = p2[y]
        soma.append((coef2, grau2))
        y = y + 1

    return soma

#Desenhar um gráfico de um polinímio

def potencia(a,b):
    res = 1
    while b > 0:
        res = res * a    
        b = b - 1
    return res

def calcPolinomio(p, x):
    res = 0
    for termo in p:
        coe, grau = termo
        res = res + coe * potencia(x,grau)
    return res

def graficoPoli(p):
    import matplotlib.pyplot as plt 
    import numpy as np
    
    x=np.arange(1,10)
    y=calcPolinomio(p,x)
    plt.plot(x,y)
    plt.show()

def criarpolinomio():
    polinomio = []
    N = int(input("Quantos termos tem o teu polinómio?"))
    while N > 0:
        n = int(input("Qual é o grau do termo"))
        a = int(input("Qual o coeficiente do termo"))
        if a != 0:
            polinomio.append((a,n))
        N = N - 1
    return polinomio

def lerlista(fnome):
    file = open(fnome, "r")
    lista = []
    for line in file:
        polinomio = line.split("\n")
        polinomio = polinomio.strip()
        for coef, grau in polinomio:
            lista.append((coef, grau))
    file.close()
    return lista

```
